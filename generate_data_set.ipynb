```python
import pandas as pd
import numpy as np
import os
import shutil

csv_file = '/share/gbpolyp/raw_data/test_label.csv'

test_original_dir = '/share/gbpolyp/raw_data/crop/test'
test_neoplastic_dir = '/share/gbpolyp/final_data/test/neoplastic'
test_nonneoplastic_dir = '/share/gbpolyp/final_data/test/nonneoplastic'

train_raw_dir = '/share/gbpolyp/raw_data/crop/train'
train_dir = '/share/gbpolyp/final_data/train'
validation_dir = '/share/gbpolyp/final_data/validation'

validation_set_ratio = 0.2
```

```python
# Test set generation

df = pd.read_csv(csv_file, dtype={'code': np.str})

if not os.path.exists(test_neoplastic_dir): os.makedirs(test_neoplastic_dir)
if not os.path.exists(test_nonneoplastic_dir): os.makedirs(test_nonneoplastic_dir)
    
for file_name in os.listdir(test_original_dir):
    file_path = os.path.join(test_original_dir, file_name)
    patient_code = file_name[0:4]
    print(patient_code)
    neoplastic = df[df['code']==patient_code].iloc[0]['neoplastic']
    
    if neoplastic == 0:
        target_path = os.path.join(test_nonneoplastic_dir, file_name)
        shutil.copy(file_path, target_path)
    else:
        target_path = os.path.join(test_neoplastic_dir, file_name)
        shutil.copy(file_path, target_path)
```

```python
# Generate train, validation set

label_list = ['neoplastic', 'nonneoplastic']

for label in label_list:
    src_dir = os.path.join(train_raw_dir, label)
    file_list = []
    for file_name in os.listdir(src_dir):
        file_list.append(os.path.join(src_dir, file_name))
    np.random.shuffle(file_list)
    
    validation_count = int(len(file_list) * validation_set_ratio)
    
    for i in range(len(file_list)):
        src_path = file_list[i]

        if i < validation_count:
            dst_dir = os.path.join(validation_dir, label)
        else:
            dst_dir = os.path.join(train_dir, label)
        
        if not os.path.exists(dst_dir): os.mkdir(dst_dir)

        dst_path = os.path.join(dst_dir, os.path.basename(src_path))
        print(src_path, dst_path)
        shutil.copy(src_path, dst_path)
```
