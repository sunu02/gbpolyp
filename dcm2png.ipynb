```python
import os
import pickle
import pydicom
import glob
from time import strptime, strftime

DEFAULT_PIXEL_SPACING = 0.145

def dcm2png(source_dirname, target_dirname):
    if os.path.basename(source_dirname) is 'error':
        return
    
    for source_filename in os.listdir(source_dirname):
        source_path = os.path.join(source_dirname, source_filename)
        
        if os.path.isfile(source_path):
            target_filename = os.path.splitext(source_filename)[0] + '.png'
            txt_filename = os.path.splitext(source_filename)[0] + '.txt'

            target_path = os.path.join(target_dirname, target_filename)
            txt_path = os.path.join(target_dirname, txt_filename)

            # dicom 2 png. Rotation되어 있는 것은 해당하는 option을 준다.
            option_dic = {'rotated90': '+on +Rl', 'rotated180': '+on +Lv', 'rotated270': '+on +Rr'}
            dir_base = os.path.basename(os.path.dirname(source_path))
            if dir_base in ['rotated90', 'rotated180', 'rotated270']:
                option = option_dic[dir_base]
            else:
                option = '+on'
            cmd = 'dcm2pnm %s %s %s' % (option, source_path, target_path)
            
            # DICOM -> PNG
            print('Converting DICOM files in %s to PNG files' % source_path)
            os.system(cmd)

            # DICOM -> txt
            # PixelSpacing이 없는 것은 default로 0.145를 출력한다.
            try:
                ds = pydicom.read_file(source_path)
                patientID = ds.PatientID
                patientSex = ds.PatientSex
                patientDOB = ds.PatientBirthDate
                studyDate = ds.StudyDate
                studyTime = ds.StudyTime
                pixelSpacing = ds.PixelSpacing
            except AttributeError as e:
                print(e)
                pixelSpacing = [DEFAULT_PIXEL_SPACING, DEFAULT_PIXEL_SPACING]
            finally:
                #print('%s %s %s %s %s' % (patientID, patientSex, patientDOB, studyDate, studyTime))
                t = strptime(studyDate[0:8] + studyTime[0:6], '%Y%m%d%H%M%S')
                studyDateTime = strftime('%Y-%m-%d %H:%M:%S', t)            
                with open(txt_path, 'w') as f:
                    f.write(source_path)                                 
                    #f.write(patientID + ',' + patientSex + ',' + patientDOB + ',' + studyDateTime)# + ',' + str(pixelSpacing[0]) + ',' + str(pixelSpacing[1])             
        else:
            dcm2png(source_path, target_dirname)
```

```python
source_dirname = '/sdb1/ybjung8594/dicom/train/adenocarcinoma'
target_dirname = '/sdb1/ybjung8594/png/train/adenocarcinoma'

dcm2png(source_dirname, target_dirname)
```
