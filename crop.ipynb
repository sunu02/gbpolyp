```python
import pydicom
import os
import glob
from scipy.misc import imread, imsave

def crop(source_dir, target_dir):
    if not os.path.exists(target_dir): os.makedirs(target_dir)
    i = 0
    for txt_file in glob.glob(os.path.join(source_dir, '*.txt')):
        source_png_path = os.path.join(source_dir, os.path.splitext(txt_file)[0] + '.png')
        try:
            with open(txt_file, 'r') as f:
                lines = f.readlines()
                if len(lines) < 2:
                    raise Exception('No crop rect or grade in "%s"' % txt_file)
                    
                split_line = lines[-1].split(',')
                x = int(float(split_line[0]))
                y = int(float(split_line[1]))
                width = int(float(split_line[2]))
                height = int(float(split_line[3]))

                crop_file_path = os.path.join(target_dir, '%s_crop.png' % os.path.splitext(os.path.split(source_png_path)[1])[0])

                if os.path.exists(crop_file_path):
                    continue

                img = imread(source_png_path, mode='L')
                cropped_img = img[y:y+height, x:x+width]
                
                imsave(crop_file_path, cropped_img)
            i = i + 1
            #if i==10: break
            print('SUCCESS>', i)
        except Exception as e:
            print('FAIL>', source_png_path, '-->', crop_file_path, ':', e)
```

```python
source_dir_list = ['/share/gbpolyp/raw_data/png/train/adenocarcinoma',
                   '/share/gbpolyp/raw_data/png/train/adenoma',
                   '/share/gbpolyp/raw_data/png/train/nonneoplastic',
                   '/share/gbpolyp/raw_data/png/test']
target_dir_list = ['/share/gbpolyp/raw_data/crop/train/adenocarcinoma',
                   '/share/gbpolyp/raw_data/crop/train/adenoma',
                   '/share/gbpolyp/raw_data/crop/train/nonneoplastic',
                   '/share/gbpolyp/raw_data/crop/test']

for source_dir, target_dir in zip(source_dir_list, target_dir_list):
    print('%s --> %s' % (source_dir, target_dir))
    crop(source_dir, target_dir)
```
