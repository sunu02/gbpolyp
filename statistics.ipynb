```python
import csv

# 0 nonneoplastaic
# 1 neoplastic

label_file = '/sdb1/ybjung8594/results/test_set.csv'
prediction_file = '/sdb1/ybjung8594/results/result3.txt'
output_file = '/sdb1/ybjung8594/results/output.csv'

label_dict = {}
with open(label_file, 'r', encoding='euc-kr') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        patient = "{0:0>4}".format(row['patient'])
        if row['nonneoplastic'] == '1':
            type = 0
        else:
            type = 1
        label_dict[patient] = type
print(label_dict)
```

```python
result_list = []
with open(prediction_file, 'r') as f:
    lines = f.readlines()
    for line in lines:
        split_line = line[:-1].split(' ')
        file_name = split_line[0]
        patient_no = split_line[0][:4]
        if split_line[1] == 'nonneoplastic':
            prediction = 0
            prob = 1 - float(split_line[2])
        else: 
            prediction = 1
            prob = float(split_line[2])
            
        label = label_dict[patient_no]
        result_list.append([file_name, patient_no, label, prediction, prob])
```

```python
n00, n01, n10, n11 = 0, 0, 0, 0
for file_name, patient_no, label, prediction, prob in result_list:
    if label == 0 and prediction == 0: n00 += 1
    elif label == 0 and prediction == 1: n01 += 1
    elif label == 1 and prediction == 0: n10 += 1
    elif label == 1 and prediction == 1: n11 += 1

print(n00, n01, n10, n11)
accuracy = (n00+n11) / (n00+n01+n10+n11)
print(accuracy)
```

```python
with open(output_file, 'w') as csvfile:
    fieldnames = ['file_name', 'label', 'prediction', 'prob']
    writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
    writer.writeheader()
    
    item = {}
    for file_name, _, label, prediction, prob in result_list:
        item['file_name'] = file_name
        item['label'] = label
        item['prediction'] = prediction
        item['prob'] = prob
        writer.writerow(item)
```
